#**우선 Machine Learning에서 중요한 개념 두가지**

* **Hypothesis**
* **Cost**

##**Hypothesis 란?**

* 개념: 가설, 즉 입력받은 값에 대한 머신의 예측값

* 형태 : H(X) = Wx + b
    
    (W = Weight(가중치), b = bias(편향 또는 성향)



###**Weight**

* 입력값에 대한 영향도를 조절하는 매개변수
  
  또는, 출력에 미치는 중요도를 조절하는 매개변수


##**Bias**

* 뉴런의 활성화를 조절하는 매개변수
  
  또는, 뉴런의 활성화 조건을 설정하는 매개변수



###**즉, Weight와 Bias를 잘 설정하여 Hypothesis(예측값)을 원래의 결과값과 비슷하게 설정하는것이 Machine Learning에서의 목표**


#**Cost란?**

* 개념: Hypothesis에서 나온 예측값(추정값)과 실제 값이 얼마나 차이가 있는지 판단하는 함수, Loss라고도 한다.
* 형태: H(X) - y이지만, 이 값이 음수가 될 수 있으므로 실제로는 제곱을 해서 구한다. => (H(x) - y) ^2
